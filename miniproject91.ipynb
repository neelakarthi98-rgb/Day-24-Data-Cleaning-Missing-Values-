## Handling Missing Values with Simple Imputation



//////Objective: Clean a dataset by identifying missing values and imputing them using
mean and mode.//

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

///Load Data://

```python
data = {
    'age': [25, 30, np.nan, 45, 29, np.nan, 38],
    'salary': [50000, 60000, 55000, np.nan, 45000, 70000, np.nan],
    'city': ['Delhi', 'Mumbai', np.nan, 'Delhi', 'Chennai', np.nan, 'Mumbai']
}
df=pd.DataFrame(data)
print("customer Data:\n",df)
```

///Identifying Missing values:

```python
miss_value=df.isnull().sum()
print("missing values in each column:\n",miss_value)
```

///Handle Numerical missing values

```python
#Check the distribution of a numerical column with missing values
#(e.g., age or salary) using a histogram
plt.Figure(figsize=(6,6))
plt.hist(df['age'].dropna())
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
# fill the missing values with mean
df['age']=df['age'].fillna(df['age'].mean())
print("Fill the missing value in age with mean:\n",df['age'])

## fill the median value into salary column
df['salary']=df['salary'].fillna(df['salary'].median())
print("Fill the missing values in salary with median :\n",df['salary'])



```

///Handle Categorical Missing Values://

```python
##Check the value counts of a categorical column with missing values 
value_count=df['city'].value_counts(dropna=False) ##we dont want to count missing values then remove "dropna=false"
print(" values in categorial column:\n",value_count)
##Fill missing values with the mode of the column:
df['city']=df['city'].fillna(df['city'].mode()[0])## we assign index=0 bez mode call too many values so we assign[0]
print("Fill the categorical coulumn with mode:\n",df['city'])
```

///Final Check:

```python
final_output=df.isnull().sum()
print("After cleaning the missing values the output is:\n",final_output)
```
