##  Categorical and Numerical Imputation

////Objective: Clean a mixed dataset with both missing numerical and
categorical values.///

```python
import pandas as pd
import numpy as np
```

///Load data://

```python
data={'age': [25, 30, np.nan, 45, 50],
    'salary': [50000, np.nan, 60000, 55000, np.nan],
    'gender': ['Male', 'Female', np.nan, 'Male', 'Female'],
    'city': ['Delhi', np.nan, 'Chennai', 'Delhi', np.nan]}
df=pd.DataFrame(data)
print("Original data is:\n",df)
d_type=df.dtypes
print("\n Data types of each column:\n",d_type)
```

///Select categorical and numerical columns//

```python
num_cols=df.select_dtypes(include=['float64']).columns
cat_cols=df.select_dtypes(include=['object']).columns
```

//Impute numerical columns///

```python
for col in num_cols:
    df[col]=df[col].fillna(df[col].median())
for col in cat_cols:
    df[col]=df[col].fillna(df[col].mode()[0])

print("\n Cleaned DtaFrame:\n",df)
```
