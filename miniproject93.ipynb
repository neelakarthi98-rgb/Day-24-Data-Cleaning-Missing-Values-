##  Time-Series Data Imputation

///Handle missing values in a time-series dataset//

```python
import numpy as np
import pandas as pd
```

////create time series dataset with missing values

```python
dates=pd.date_range(start='2025-11-01',periods=10,freq='D')
data=[10,20,np.nan,25,np.nan,15,10,np.nan,30,32]
df=pd.DataFrame({"date":dates,"value":data})
print(df)


```

///Set the date column as index//

```python
df['date'] = pd.to_datetime(df['date']) ##convert into datetime index
df=df.set_index('date')
print("Original Dataframe with missing values\n")
print(df)
```

///Handle missing values using time-based interpolation//

```python
df_interpolated = df.interpolate(method='time')

print("\nDataFrame after time-based interpolation:")
print(df_interpolated)
```
